```python
import pandas as pd
```

```python
df_top = pd.read_csv("sustainalytics_scores.csv", index_col=False)
df_top
```

```python
df_bottom = pd.read_csv("sustainalytics_scores_bottom.csv", index_col=False)
df_bottom
```

```python
sustain_df = pd.concat([df_top, df_bottom])
sustain_df
```

```python
sustain_df_filtered = sustain_df.drop_duplicates(subset=['exchange:ticker'], keep='first')
sustain_df_filtered
```

```python
sustain_df_filtered[['exchange', 'ticker']] = sustain_df_filtered['exchange:ticker'].str.split(':', expand=True)
sustain_df_filtered
```

```python
us_stocks = pd.read_csv("us_stocks.csv")
us_stocks
```

```python
sustain_final = pd.merge(sustain_df_filtered, 
                     us_stocks, 
                     on ='ticker', 
                     how ='left')
sustain_final
```

```python
sustain_final = sustain_final.drop('exchange:ticker', axis=1)
sustain_final = sustain_final.drop('exchange_x', axis=1)
sustain_final = sustain_final.drop('is_etf', axis=1)
sustain_final
```

```python
sustain_final = sustain_final.rename(columns={'exchange_y':'exchange'})
sustain_final
```

```python
sustain_final = sustain_final[['name', 'ticker', 'exchange','Sustainalytics_score']]
sustain_final
```

```python
sustain_final.to_csv("sustainalytics_scores_final.csv",index=False)
```
